```python
# Importing OpenCV and Numpy Libraries
import cv2
import numpy as np
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image

from PIL import Image

import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib.pyplot import rcParams
rcParams['figure.figsize'] = 12,12
```

```python
img = cv2.imread('lena.jpeg')
#img = cv2.imread('example.png')

#img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#img = img[0]*0.33 + img[1]*0.33 + img[2]*0.33
```

```python
slp33 = np.array([[0.1111111, 0.1111111, 0.1111111],
                    [0.1111111, 0.1111111, 0.1111111],
                    [0.1111111, 0.1111111, 0.1111111]])
  
lp33 = cv2.filter2D(src=img, ddepth=-1, kernel=slp33)

slp55 = np.array([[0.04,0.04,0.04,0.04,0.04],
                 [0.04,0.04,0.04,0.04,0.04],
                 [0.04,0.04,0.04,0.04,0.04],
                 [0.04,0.04,0.04,0.04,0.04],
                 [0.04,0.04,0.04,0.04,0.04]])
  
lp55 = cv2.filter2D(src=img, ddepth=-1, kernel=slp55)

gaussian_blur = cv2.GaussianBlur(src=img, ksize=(5,5),sigmaX=0, sigmaY=0)
```

```python
fig = plt.figure(figsize=(15, 10))

# Orginial Image
fig.add_subplot(2, 2, 1)
plt.imshow(img)
plt.title("Original Image")

# 3x3 spatial low pass Filtered Image
fig.add_subplot(2, 2, 2)
plt.imshow(lp33)
plt.title("3x3 Spatial filtered Image")

# 3x3 spatial low pass Filtered Image
fig.add_subplot(2, 2, 3)
plt.imshow(lp55)
plt.title("5x5 Spatial filtered Image")

# 3x3 spatial low pass Filtered Image
fig.add_subplot(2, 2, 4)
plt.imshow(gaussian_blur)
plt.title("3x3 guassian Spatial filtered Image")
```
